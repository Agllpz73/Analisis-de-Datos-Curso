Limpieza de datos

En el analisis de datos se necesita estar haciendo una limpieza para poder visualizar los datos m√°s facilmente
Por lo que tenemos diferentes forma de hacer dicha limpieza

```python
#Limpieza de datos

df_filtrado = df.dropna()
df_filtrado.head()
```

Iniciemos primero con una funci√≥n de filtro/limpieza que es la funci√≥n .dropna()

Sintaxis:
dataframe.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)

Donde:
axis: Especifica si se eliminan filas (0) o columnas (1). Por defecto es 0, es decir, filas.
how: Define si se eliminan las filas o columnas que contienen cualquier valor nulo ('any') o solo si todas las entradas son nulas ('all').
thresh: Indica el n√∫mero m√≠nimo de valores no nulos que deben estar presentes para que la fila o columna no sea eliminada.
subset: Permite especificar un subconjunto de columnas para la evaluaci√≥n.
inplace: Si es True, modifica el DataFrame original en lugar de devolver una copia del DataFrame sin los valores nulos.

```python

df_filtrado = df.fillna(0)
df_filtrado.head()
```

Ahora tenemos la funci√≥n .fill( valor a reemplazar)

Esta funci√≥n, a comparaci√≥n que .dropna() hace que sustituye cada elemento "vacio" del dataframe por el valor que deseemos, esto es √∫til cuando queremos mantener las columnas o filas vacias para poder reutilizar o llenarlas m√°s adelante


```python
df_filtrado = df.fillna({"retweets" : 0, "mentions": -1})
df_filtrado.head()
```

COMO FILTRAR DATOS CON CONDICIONES

Los dataframe como tal se pueden filtrar con base en condiciones, ya que como desconocemos los datos que contiene esto suele ser complicado de ver, por lo tanto, podemos aplicar condicionales para poder tener una mayor visualizaci√≥n de los dato.

Sintaxis:

dataframe[dataframe["nobre_columna"] - y la condici√≥n que desees]

La condici√≥n depender√° del tipo de dato que contega la columna del dataframe, se pueden aplicar condicionales aritm√©ticas o incluso para el manejo de String's, etc.

Nota: Es importante destacar que con esta sintaxis nosotros podemos concatenar diversas condiciones

```python
#Ejemplo


df[(df["favorites"]>400) & (df["retweets"]==24)] 
```

```python
#Ejemplo con cadena de caracteres o String's

df[df["full_text"].str.contains("Programming")]
```

Fase de transformaci√≥n de datos

Se pueden transformar datos dentro de un dataframe, esto se suele usar cuando queremos agregar una columna con datos extras que hasta cierto punto consideramos utiles al momento de analisar los datos.

SINTAXIS:

df["nombre_columna_nueva"] = df["nombre_columna_usar"].apply(funcion)


Algo importante a destacar es que esta sintaxis es para poder crear una nueva columna dentro del dataframe. La palabra funcion se refiere a la funci√≥n a la que deseeamos aplicar al dataframe con respecto a la columna a modificar, etc.

USAR DOS O M√ÅS COLUMNAS PARA AGREGAR UNA COLUMNA EXTRA

En ocasiones y en la mayoria de los casos requerimos hacer uso de dos o m√°s columnas del dataframe, por lo que, se puede hacer siguiendo esta sintaxisis


SINTAXIS:

df["nombre_columna_nueva"] = df.apply(funcion, axis=1)


Parecida a la sintaxis cuando manipulas una sola funci√≥n, la diferencia aqu√≠ es que no estamos indicando la columna a la que quiero que se aplique la funci√≥n, sino que estoy indicando de  igual forma que la funci√≥n se aplique a todo el dataframe, sin embargo, dentro de la funci√≥n indico "implicitamente" a que columnas voy agarrar.

Algo importante de esta sintaxis es "axis" ya que este especifica a .apply() que la funci√≥n deber ser aplicada por cada fila del dataframe.

```python
#EJEMPLO
import random

def calcularGanancias(retweets):
    ganancias = retweets*random.randint(3,5)
    return ganancias


df["ganancias"] = df["retweets"].apply(calcularGanancias)

df.head(10)
```

```python
#Ejemplo cuando quiero manipular dos o m√°s columnas
import random

def popularidad(fila):
    
    resultado = fila["followees"] / fila["followers"]
    return resultado

df["popularidad"] = df.apply(popularidad, axis=1)

df.head()
```

AGRUPACI√ìN DE DATOS

Definici√≥n: Agrupar significa unir filas que tengan un mismo valor dentro de una columna. Las agrupaciones nos ayudan a poder hacer caluculos m√°s espec√≠ficos, por ejemplo, de una regi√≥n o pa√≠s.

SINTAXIS

dataframe.groupby("nombre_columna").funcionPythonParaAplicar(......)

Con esta sintaxis gen√©rica podemos generar calculas que nosotros deseemos


```python

df.groupby("country").agg({"followers" : 'sum', "mentions" : 'mean', "retweets" : 'max'})
```

```python
df.groupby("country").sum()
```

```python
#nota importante, esta funcion de agrupaci√≥n genera un nuevo dataframe al que le podemo aplicar cualquier otra acci√≥n


group = df.groupby("country").agg({"followers" : 'sum', "mentions" : 'mean', "retweets" : 'max'}) #esto es un dataframe


group[group["followers"] < 5000]
```
