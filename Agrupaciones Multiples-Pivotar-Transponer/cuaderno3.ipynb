### FECHAS

Por defecto, cuando se crea un DataFrame los campos fechas suelen ser de tipo String, por lo tanto se pueden cambiar a un formato de fecha como se muestra a continuaci√≥n

```python
df_completo['fecha'] = pd.to_datetime(df_completo['fecha_publicado'], format="%d/%m/%Y") ## Importante de no dejar espacios ya que pandas espera que est√© exactamente como esta

df_completo
```

```python
# Obtener d√≠a de la semana de un campo de tipo fecha

df_completo['dia_semana'] = df_completo['fecha'].apply(lambda x : x.strftime('%A'))
```
