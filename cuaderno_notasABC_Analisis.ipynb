Limpieza de datos

En el analisis de datos se necesita estar haciendo una limpieza para poder visualizar los datos m√°s facilmente
Por lo que tenemos diferentes forma de hacer dicha limpieza

```python
#Limpieza de datos

df_filtrado = df.dropna()
df_filtrado.head()
```

Iniciemos primero con una funci√≥n de filtro/limpieza que es la funci√≥n .dropna()

Sintaxis:
dataframe.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)

Donde:
axis: Especifica si se eliminan filas (0) o columnas (1). Por defecto es 0, es decir, filas.
how: Define si se eliminan las filas o columnas que contienen cualquier valor nulo ('any') o solo si todas las entradas son nulas ('all').
thresh: Indica el n√∫mero m√≠nimo de valores no nulos que deben estar presentes para que la fila o columna no sea eliminada.
subset: Permite especificar un subconjunto de columnas para la evaluaci√≥n.
inplace: Si es True, modifica el DataFrame original en lugar de devolver una copia del DataFrame sin los valores nulos.

```python

df_filtrado = df.fillna(0)
df_filtrado.head()
```

Ahora tenemos la funci√≥n .fill( valor a reemplazar)

Esta funci√≥n, a comparaci√≥n que .dropna() hace que sustituye cada elemento "vacio" del dataframe por el valor que deseemos, esto es √∫til cuando queremos mantener las columnas o filas vacias para poder reutilizar o llenarlas m√°s adelante


```python
df_filtrado = df.fillna({"retweets" : 0, "mentions": -1})
df_filtrado.head()
```

Filtrado por filas

En ocasiones no ser√° necesario filtrar por columnas sino por filas, esto suele llegar a ser m√°s practico si solamente queremos analizar datos en espec√≠ficos


Funci√≥n .iloc[]

Esta funci√≥n se dedica a regresarnos la fila completa que nostros le indiquemos, sin embargo, es bueno se√±alar que necesitamos conocer el indice de la fila que desemos, es decir, recordemos que cada fila est√° enumerado por un √≠ndice dado que va desde el 0 hasta N, donde N es el √∫ltimo elemento que tenemos

Sintaxis:
dataframe.iloc[indice_fila]

De igual forma podemos generar un rango de filas para poder analizarlas

Sintaxis:

dataframe.iloc[i:N]

Tengamos cuidado, ya que este √∫ltimo rango de filas se debe considerar que el rango de filas solamente llegar√° hasta N-1 fila en donde N es el m√°ximo de filas a las que llegaremos, es decir, si N=3, el rango de filas llegar√° hasta la fila 2, RECORDEMOS que se inicia en 0, como en otro tipo de estructura de datos


```python
df.iloc[0]
```

Cabe mencionar que tambi√©n podemos hacer filtrados por indicadores de fila, usando la funci√≥n .loc[]

Sintaxis:

dataframe.loc[valor_identificador]

dataframe.loc[['identificador_1', 'identificador_2', .... , 'identificador_N]]  ---> esto tambi√©n se puede hacer mediante la funci√≥n .iloc[], hacer filtrado de filas en espec√≠fico
